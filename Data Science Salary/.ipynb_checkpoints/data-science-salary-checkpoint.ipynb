## STATES WITH MOST NUMBER OF JOBS

```python
# Label names for the state
country = ["California","Massachusetts","New York","Virginia","Illinois","Maryland","Pennsylvania","Texas","Washington","North Carolina"]
from IPython.core.pylabtools import figsize

# Top 10 states
fig, ax = plt.subplots(nrows=1, ncols=1)
a = sns.barplot(x=df["Job Location"].value_counts().index[0:10], y=df["Job Location"].value_counts()[0:10])

# Remove top and right borders
sns.despine(bottom=False, left=False)

import matplotlib
from matplotlib import rcParams
rcParams['figure.figsize'] = 16,8

# Add % on the bar plot
# Stackoverflow
l = df["Job Location"].value_counts().index[0:10]
for p in ax.patches:
    ax.text(p.get_x() + 0.1, p.get_height()+4.5, '{:.2f}%'.format((p.get_height()/742)*100))
    
# Add nice visualization
plt.title('\n States with Most Number of Jobs \n', size=24, color='black')
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel('\n States \n', fontsize=18, color='black')
plt.ylabel('\n Count \n', fontsize=18, color='black')
patches = [matplotlib.patches.Patch(color=sns.color_palette()[i], label=j) for i,j in zip(range(0,10),country)]
plt.legend(handles=patches, loc="upper right")
plt.show()
```

## TOP 5 INDUSTRIES HIGHEST NUMBER OF DATA SCIENCE RELATED JOB POSTING

```python
df["Industry"].nunique() # number of industries
```

```python
# Use pie chart for the visualization

divisions=(0,0.1,0.1,0.1,0.1)
labels= [x for x in df["Industry"].value_counts().sort_values(ascending=False)[0:5].index]
patches,ax, text= plt.pie(df["Industry"].value_counts().sort_values(ascending=False)[0:5],autopct='%1.1f%%',explode=divisions,shadow=True,startangle=305)
plt.title('\n Top 5 Industries with Most Number of Data Science Related Jobs \n', size=20, color='black')
plt.legend(patches, labels, loc="best")
plt.axis('equal')
plt.show()
```

## JOB TITLES WITH MOST NUMBER OF JOBS

```python
# Demanded job titles
job_t = sns.catplot(x="job_title_sim",data=df,order=df.job_title_sim.value_counts().index,kind="count")
job_t.fig.set_size_inches(10,5)

#Putting % on the bar plot
divisions = df.job_title_sim.value_counts().index
ax = job_t.facet_axis(0,0)
for p in ax.patches:
    ax.text(p.get_x() + 0.15, p.get_height()+5, '{:.2f}%'.format((p.get_height()/742)*100))

#Beautifying the plot
plt.title('\n Job Titles \n', size=24, color='black')
plt.xticks(rotation=50,ha='right', rotation_mode='anchor')
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
plt.xlabel('\n Job Title \n', fontsize=18, color='black')
plt.ylabel('\n Count \n', fontsize=18, color='black')
plt.show()
```
